```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.gridspec import GridSpecFromSubplotSpec 
```

```python
figsize = None
n_rows = 3
n_cols = 3
n_results = 8

_, axs = plt.subplots()
axs.set_axis_off()
axes_ = np.empty((n_rows, n_cols), dtype=object)
figure_ = axs.figure
if figsize is None:
    figure_.set_size_inches(n_cols * 7, n_rows * 5)
else:
    figure_.set_size_inches(figsize[0], figsize[1])
axes_ravel = axes_.ravel()

gs = GridSpecFromSubplotSpec(
    n_rows, n_cols, subplot_spec=axs.get_subplotspec()
)

for i, spec in zip(range(n_results), gs):
    axes_ravel[i] = figure_.add_subplot(spec)

for i in range(n_results):
    ax = axes_ravel[i]
    ax.scatter(np.arange(1, 10), np.arange(1, 10))
    # for j in ax.spines.values():
    #     j.set_linewidth(1)
    plt.setp(ax.spines.values(), linewidth=1)

# for ax_spines in axes_ravel.spines.values():
#     ax_spines.set_linewidth(10)
```

```python
def test_subplots():
    figsize = None
    n_rows = 3
    n_cols = 3
    n_results = 8

    _, axs = plt.subplots()
    axs.set_axis_off()
    axes_ = np.empty((n_rows, n_cols), dtype=object)
    figure_ = axs.figure
    if figsize is None:
        figure_.set_size_inches(n_cols * 7, n_rows * 5)
    else:
        figure_.set_size_inches(figsize[0], figsize[1])
    axes_ravel = axes_.ravel()

    gs = GridSpecFromSubplotSpec(
        n_rows, n_cols, subplot_spec=axs.get_subplotspec()
    )

    for i, spec in zip(range(n_results), gs):
        axes_ravel[i] = figure_.add_subplot(spec)

    for i in range(n_results):
        ax = axes_ravel[i]
        ax.scatter(np.arange(1, 10), np.arange(1, 10))
        # for j in ax.spines.values():
        #     j.set_linewidth(1)
        plt.setp(ax.spines.values(), linewidth=1)
    return figure_
```
